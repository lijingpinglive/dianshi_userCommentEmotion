```python
import pandas as pd

data_train = pd.read_csv('data_train.csv',sep='\t',header = None,names=['id','theme','content','label'],encoding='gbk')
data_train.head(5)
```

```python
#转换成字符串格式
content = data_train.content.values.astype(str)
#将数组转换成列表
content = content.tolist()
print(content[0])
type(content)
```

```python
import jieba
content_cut = []
#jieba.lcut(content[0])

for line in content:
    content_cut.append(jieba.lcut(line))
```

```python
content_cut[0]
```

```python
word_dict ={}
test = content_cut[0]
for word in test:
    #print(word)
    if word not in word_dict:
        word_dict[word]=1
    else:
        word_dict[word] += 1
#print(word_dict) 
```

```python
word_all = {}
for word_1 in content_cut:
    for word_2 in word_1:
        if word_2 not in word_all:
            word_all[word_2] = 1
        else:
            word_all[word_2] += 1
          

        
#print(word_all)
word_all = sorted(word_all.items(),key = lambda x:x[1],reverse = True)
print(word_all)
```

```python
stopwords = pd.read_csv('mystopwords.txt',sep='\t',encoding='utf-8',header=None, names=['stopwords'])
stopwords = stopwords.stopwords.values.tolist()
```

```python
#content_df = pd.DataFrame({"content_cut":content_cut})
#content_df.head()
```

```python
def stopwordsClen(contents, stopwords):
    contents_clean = []
    for line in contents:
        line_clean = []
        for word in line:
            if word  in stopwords:
                continue
            line_clean.append(word)
        contents_clean.append(line_clean)
    return contents_clean
```

```python
contents_clean = stopwordsClen(content_cut,stopwords)
contents_clean[0]
```

```python
df_train = pd.DataFrame({"contents_clen":contents_clean , "label":data_train['label']})
df_train.tail()
```

```python
#将数据拆分成测试集和训练集
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df_train['contents_clen'].values,df_train['label'].values, random_state=1)
```

```python
x_train.shape
x_train[0]
```

```python
y_train.shape
y_train[0]
```

```python
words = []
for line_index in range(len(x_train)):
    words.append(" ".join(x_train[line_index]))
words[0]

words_test = []
for line_index in range(len(x_test)):
    words_test.append(" ".join(x_test[line_index]))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector = TfidfVectorizer(analyzer='word', max_features=4000,  lowercase = False)
```

```python
from sklearn.naive_bayes import MultinomialNB
#vector.fit(words)
classifier = MultinomialNB()
classifier.fit(vector.transform(words), y_train)
```

```python
classifier.score(vector.transform(words_test), y_test) 
```

```python
#进行预测
data_test = pd.read_csv('./data/data_test.csv',sep='\t',header = None,names=['id','theme','content'],encoding='gbk')
content_pre = data_test.content.values.astype(str)
#将数组转换成列表
content_pre = content_pre.tolist()
content_pre_cut = []
for i in content_pre:
    content_pre_cut.append(jieba.lcut(i))
```

```python
contents_pre_clean = stopwordsClen(content_pre_cut,stopwords)
contents_pre_clean[0]
```

```python
words_pre = []
for line_index in range(len(content_pre_cut)):
    words_pre.append(" ".join(contents_pre_clean[line_index]))

words_pre[0]
```

```python
result = classifier.predict(vector.transform(words_pre))
```

```python
result
```

```python
result_pre = pd.DataFrame({"id":data_test['id'] , 'label':result})
result_pre.head()
```
